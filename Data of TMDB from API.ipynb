```python
# code for api data collection

final_df = pd.DataFrame(columns=['title', 'overview', 'vote_count', 'original_language', 'release_date', 'popularity', 'vote_average'])

for page in range(1,500):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=85ec53122b9821b68d3c017559ba0260&language=en-US&page={}'.format(page)).json()
    
    popularity=[]
    vote_count=[]
    original_language=[]
    title=[]
    vote_average=[]
    overview=[]
    release_date=[]
    for movie in response['results']:
        
        try:
            popularity.append(movie['title'])
        except:
            popularity.append('')
            
        try:
            vote_count.append(movie['vote_count'])
        except:
            vote_count.append('')
            
        try:
            original_language.append(movie['original_language'])
        except:
            original_language.append('')
            
        try:
            title.append(movie['title'])
        except:
            title.append('')
            
        try:
            vote_average.append(movie['vote_average'])
        except:
            vote_average.append('')
            
        try:
            overview.append(movie['overview'])
        except:
            overview.append('')
            
        try:
            release_date.append(movie['release_date'])
        except:
            release_date.append('')
        

    dict_of_data = {'title':title, 'overview':overview,'vote_count':vote_count, 'original_language':original_language,'release_date':release_date, 'popularity':popularity, 'vote_average':vote_average}

    df = pd.DataFrame(dict_of_data)  
    final_df = pd.concat([final_df, df], ignore_index = True)
    
final_df
```

```python
final_df.to_csv('TMDB_movies_data.csv', index=False)
```
